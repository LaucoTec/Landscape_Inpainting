## Primer intento: Simple reconstrucción

### Cargar datasets

```python
trainFolder = '/content/Inpainting_TC3007C/paisajes/Landscape Classification/Landscape Classification/Training Data/Mountain'
valFolder = '/content/Inpainting_TC3007C/paisajes/Landscape Classification/Landscape Classification/Validation Data/Mountain'
imgSize = (290, 174) # Tamaño original de las imágenes
batchSize = 32 # Tamaño de bloques

trainDS = tf.keras.utils.image_dataset_from_directory(
    trainFolder,
    labels=None,
    label_mode=None,
    image_size=imgSize,
    batch_size=batchSize,
    shuffle=True
)

valDS = tf.keras.utils.image_dataset_from_directory(
    valFolder,
    labels=None,
    label_mode=None,
    image_size=imgSize,
    batch_size=batchSize,
    shuffle=True
)

print(f"Dataset de entrenamiento cargado desde {trainFolder}")
print(f"Dataset de validación cargado desde {valFolder}")
```

Cargar únicamente conjuntos para entrenar y validar. no hay intención de seguir trabajando sobre esta pequeña iteración para probar más a detalle con un set dedicado de prueba.

La propia imagen es el objetivo, por lo que no se extraen etiquetas. Se separa por batches de 32 imágenes con mezclado aleatorio para que no se memorice el orden de los datos originales.

### Preprocesamiento

```python
# Preprocesamiento a imágenes
def trainPre(image):
    image = tf.image.resize(image, (128, 128))
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.image.random_flip_left_right(image)
    return image, image

def valPre(image):
    image = tf.image.resize(image, (128, 128))
    image = tf.cast(image, tf.float32) / 255.0
    return image, image


# Preprocesamiento a datasets
trainDS = (
    trainDS
    .shuffle(1000)
    .map(trainPre, num_parallel_calls=tf.data.AUTOTUNE)
    .cache()
    .prefetch(tf.data.AUTOTUNE)
)

valDS = (
    valDS
    .map(valPre, num_parallel_calls=tf.data.AUTOTUNE)
    .cache()
    .prefetch(tf.data.AUTOTUNE)
)


print(f"Nuevos detalles de entrenamiento: {trainDS.element_spec}")
print(f"Nuevos detalles de validación: {valDS.element_spec}")
```

En el preprocesado se normalizan todas las imágenes para valores de entre 0 y 1 para facilitar la convergencia, se activa un barajeo de mil imágenes para que en las épocas no se memorice un orden fijo ahora de los datos en el set cargado. Cache y prefetch permiten cargar todo en memoria aprovechando que aún son pocos datos para apurar el entrenamiento. Las imágenes se reducen a un 128x128 pues no se ocupa tanta información para la única clase que debe aprender.

Finalmente, el único aumento de datos aplicado es una inversión horizontal de las imágenes.  

### Construir el modelo

```python
# Profundidad de 3
encoder = keras.Sequential([
    layers.Input(shape=(128, 128, 3)),
    layers.Conv2D(16, 3, activation='relu', strides=2, padding='same'),
    layers.BatchNormalization(),

    layers.Conv2D(32, 3, activation='relu', strides=2, padding='same'),
    layers.BatchNormalization(),

    layers.Conv2D(64, 3, activation='relu', strides=2, padding='same'),
    layers.BatchNormalization(),

    layers.Flatten(),
    layers.Dense(1024, activation='linear')
])

decoder = keras.Sequential([
    layers.Input(shape=(1024,)),

    layers.Dense(16 * 16 * 64, activation='relu'),
    layers.Reshape((16, 16, 64)),

    layers.UpSampling2D(size=2),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),

    layers.UpSampling2D(size=2),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),

    layers.UpSampling2D(size=2),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),

    layers.Conv2D(3, 3, padding='same', activation='sigmoid')
])
```

La arquitectura es sencilla con una profundidad de 3 antes del cuello de botella:
- Convoluciones con 2 de padding para reducir el tamaño y filtros de 16 hasta 64 para algo ligero.
  - ReLu se usa como función de activación aprovechando el recorte de 0 a 1 y evitar el desvanecimiento de gradiente.
- La regularización se hace por bloques al tener un tamaño de batch no tan pequeño, tras cada convolución.
- El cuello de botella final es una capa densa de 1024 neuronas de activación lineal.
- El decodificador, con la operación opuesta, realiza el upsampling antes de las convoluciones.
- La salida final es con una convolución de 3 filtros y kernel de 3.
  - Su activación es sigmoide ya que los valores originales se encontraban ya entre 0 y 1.

```python
# Definir la entrada
input_img = keras.Input(shape=(128, 128, 3))

# Generar el encoder
encoded = encoder(input_img)

# Generar el decoder
decoded = decoder(encoded)

# Unir las partes
autoencoder = keras.Model(inputs=input_img, outputs=decoded)
autoencoder.summary()
```

```python
# Generar modelo final
autoencoder.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss=keras.losses.MeanSquaredError()
)

# Definir callbacks durante entrenamiento
# Early Stop
early = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)
# Checkpoints
ckptPath = "/content/drive/MyDrive/Colab Notebooks/Inpainting/Models/BasicAE/ae_best.h5"
ckpt = keras.callbacks.ModelCheckpoint(
    filepath=ckptPath,
    monitor="val_loss",
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)
```

Adam se usa como optimizador frecuentemente gracias a que puede ajustar la tasa de aprendizaje y corregir sesgos mientras mantiene un gran desempeño. La tasa original de 0.001 es de igual forma un valor frecuente.

Se definen dos callbacks:
- EarlyStopping para prevenir sobreajuste cuando la pérdida del set de validación ya no mejore o empeore.
- Checkpoint para guardar la mejor versión del modelo siempre.

```python
history = autoencoder.fit(
    trainDS,
    epochs=50,
    validation_data=valDS,
    callbacks=[early, ckpt]
)
```

## Tercer intento: Camino al Inpainting

### Cargar sets de datos

```python
trainFolder = '/content/Inpainting_TC3007C/paisajes/Training Data'
valFolder = '/content/Inpainting_TC3007C/paisajes/Validation Data'
testFolder = '/content/Inpainting_TC3007C/paisajes/Testing Data'
imgSize = (160, 272) # Tamaño original: 174x290
batchSize = 32

trainDS = tf.keras.utils.image_dataset_from_directory(
    trainFolder,
    labels=None,
    label_mode=None,
    image_size=imgSize,
    batch_size=None,
    shuffle=True
)

valDS = tf.keras.utils.image_dataset_from_directory(
    valFolder,
    labels=None,
    label_mode=None,
    image_size=imgSize,
    batch_size=None,
    shuffle=True
)

testDS = tf.keras.utils.image_dataset_from_directory(
    testFolder,
    labels=None,
    label_mode=None,
    image_size=imgSize,
    batch_size=None,
    shuffle=True
)

print(f"Dataset de entrenamiento cargado desde {trainFolder}")
print(f"Con forma {trainDS.element_spec}")
print(f"Dataset de validación cargado desde {valFolder}")
print(f"Con forma {valDS.element_spec}")
print(f"Dataset de prueba cargado desde {testFolder}")
print(f"Con forma {testDS.element_spec}")
```

### Preprocesamiento

```python
# Generación de máscaras al momento (estilo pincelazo)
def genMask(num_strokes=10, max_brush_size=30):
    # Convert the TensorFlow tensor to Python integers
    height = imgSize[0]
    width = imgSize[1]

    color = 255
    mask = np.zeros((height, width), dtype=np.uint8) # Color negro

    for i in range(num_strokes):
        # Punto inicial aleatorio
        x1 = random.randint(0, width)
        y1 = random.randint(0, height)

        # Punto final aleatorio
        x2 = random.randint(0, width)
        y2 = random.randint(0, height)

        # Grosor aleatorio
        thickness = random.randint(5, max_brush_size)

        # Dibuja toda la línea o sólo círculo aleatoriamente
        if random.random() > 0.5:
            cv2.line(mask, (x1, y1), (x2, y2), color, thickness)
        else:
            radius = thickness // 2
            cv2.circle(mask, (x1, y1), radius, color, -1)

    return mask.astype(np.float32) / 255.0
```

```python
# Preprocesamiento a imágenes
def maskTrain(image, label=None):
    image = tf.cast(image, tf.float32) / 255.0

    maskTF = tf.py_function(genMask, [], tf.float32)
    maskTF.set_shape((imgSize[0], imgSize[1]))
    maskTF = tf.expand_dims(maskTF, axis=-1)
    maskTF3 = tf.repeat(maskTF, repeats=3, axis=-1)

    maskedImg = image * (1-maskTF3)
    target = tf.concat([image, maskTF3], axis=-1)

    return tf.cast(maskedImg, tf.float16), tf.cast(target, tf.float16)

def valTestPre(image, label=None):
    image = tf.cast(image, tf.float32) / 255.0
    image.set_shape((imgSize[0], imgSize[1], 3))

    maskTF = tf.py_function(genMask, [], tf.float32)
    maskTF.set_shape((imgSize[0], imgSize[1]))
    maskTF = tf.expand_dims(maskTF, axis=-1)
    maskTF3 = tf.repeat(maskTF, repeats=3, axis=-1)

    maskedImg = image * (1-maskTF3)
    target = tf.concat([image, maskTF3], axis=-1)

    return maskedImg, target


# Preprocesamiento a datasets
trainDS = (
    trainDS
    .shuffle(1000)
    .map(maskTrain, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(batchSize)
    .prefetch(tf.data.AUTOTUNE)
)

valDS = (
    valDS
    .map(valTestPre, num_parallel_calls=tf.data.AUTOTUNE)
    .cache()
    .batch(batchSize)
    .prefetch(tf.data.AUTOTUNE)
)

testDS = (
    testDS
    .map(valTestPre, num_parallel_calls=tf.data.AUTOTUNE)
    .cache()
    .batch(batchSize)
    .prefetch(tf.data.AUTOTUNE)
)


print(f"Nuevos detalles de entrenamiento: {trainDS.element_spec}")
print(f"Nuevos detalles de validación: {valDS.element_spec}")
print(f"Nuevos detalles de prueba: {testDS.element_spec}")
```

### Definición del modelo

```python
# Función de pérdida
def charbonnierLoss(y_true, y_pred, mask, epsilon=3e-3):
    # Error sólo en la máscara
    diff = y_true - y_pred
    totalError = tf.sqrt(tf.square(diff) + tf.square(epsilon))
    maskedError = totalError * mask

    # Normalizar pérdida
    numNormal = tf.reduce_sum(mask) + 1e-6 # Evitar división entre 0
    return tf.reduce_sum(maskedError) / numNormal

lpip = lpips_base_tf.LPIPS(base='vgg16', pre_norm=True)
def lpipsLoss(y_true, y_pred):
  # LPIPS sí se calcula sobre la imagen entera
  return lpip(y_true, y_pred)

def combLoss(y_true, y_pred):
  target = y_true[..., :3]
  mask = y_true[..., 3:]

  charbonnierMask = charbonnierLoss(target, y_pred, mask)

  lpipsFull = lpipsLoss(target, y_pred)

  return charbonnierMask + 0.2 * lpipsFull
```

```python
# ---Arquitectura---
# Entrada
inp = keras.Input(shape=(160, 272, 3))

# Encoder
e1 = layers.Conv2D(32, 7, activation='relu', strides=2, padding='same')(inp)
e1 = layers.GroupNormalization(groups=1)(e1)

e2 = layers.Conv2D(64, 3, activation='relu', strides=2, padding='same')(e1)
e2 = layers.GroupNormalization(groups=1)(e2)

e3 = layers.Conv2D(128, 3, activation='relu', strides=2, padding='same')(e2)
e3 = layers.GroupNormalization(groups=1)(e3)

e4 = layers.Conv2D(256, 3, activation='relu', padding='same')(e3)
e4 = layers.GroupNormalization(groups=1)(e4)

# Cuello de botella
bottleneck = layers.Conv2D(512, 3, activation='relu', padding='same', dilation_rate=2)(e4)
bottleneck = layers.GroupNormalization(groups=1)(bottleneck)
bottleneck = layers.Conv2D(512, 3, activation='relu', padding='same')(bottleneck)
bottleneck = layers.GroupNormalization(groups=1)(bottleneck)

# Decoder
d4 = layers.Conv2DTranspose(256, 3, activation='relu', padding='same', dilation_rate=2)(bottleneck)
d4 = layers.GroupNormalization(groups=1)(d4)
d4 = layers.Concatenate()([d4, e3])

d3 = layers.Conv2DTranspose(128, 3, activation='relu', strides=2, padding='same')(d4)
d3 = layers.GroupNormalization(groups=1)(d3)
d3 = layers.Concatenate()([d3, e2])

d2 = layers.Conv2DTranspose(64, 3, activation='relu', strides=2, padding='same')(d3)
d2 = layers.GroupNormalization(groups=1)(d2)
d2 = layers.Concatenate()([d2, e1])

d1 = layers.Conv2DTranspose(32, 7, activation='relu', strides=2, padding='same')(d2)
d1 = layers.GroupNormalization(groups=1)(d1)

# Salida
out = layers.Conv2D(3, 3, activation='sigmoid', padding='same')(d1)

# Modelo
autoencoder = keras.Model(inputs=inp, outputs=out)
```

```python
# Compilar modelo
autoencoder.compile(
    optimizer=keras.optimizers.Adam(learning_rate=2e-3),
    loss= combLoss
)

# ---Callbacks de entrenamiento---
# Early Stop
early = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)
# Checkpoints
ckptPath = "/content/drive/MyDrive/Colab Notebooks/Inpainting/Models/InpaintingAE/ip3_best.keras"
ckpt = keras.callbacks.ModelCheckpoint(
    filepath=ckptPath,
    monitor="val_loss",
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)
# Learning rate scheduler
lrScheduler = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    verbose=1,
    min_lr=1e-4
)
```

### Entrenamiento

```python
# Entrenar el modelo
history = autoencoder.fit(
    trainDS,
    epochs=60,
    validation_data=valDS,
    callbacks=[early, ckpt, lrScheduler]
)
```

### Evaluación

```python
# Importaciones requeridas
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import random

# Generación de máscaras aleatorias
def genMask(num_strokes=10, max_brush_size=30):
    # Convert the TensorFlow tensor to Python integers
    height = imgSize[0]
    width = imgSize[1]

    color = 255
    mask = np.zeros((height, width), dtype=np.uint8) # Color negro

    for i in range(num_strokes):
        # Punto inicial aleatorio
        x1 = random.randint(0, width)
        y1 = random.randint(0, height)

        # Punto final aleatorio
        x2 = random.randint(0, width)
        y2 = random.randint(0, height)

        # Grosor aleatorio
        thickness = random.randint(5, max_brush_size)

        # Dibuja toda la línea o sólo círculo aleatoriamente
        if random.random() > 0.5:
            cv2.line(mask, (x1, y1), (x2, y2), color, thickness)
        else:
            radius = thickness // 2
            cv2.circle(mask, (x1, y1), radius, color, -1)

    return mask.astype(np.float32) / 255.0
```
